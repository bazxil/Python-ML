```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Load_Extension - Sheet1-checkpoint.csv")
```

```python
print(data)
```

```python
x = data[["Load (N)"]]
y = data[["Extension (mm)"]]
```

```python
#print(x)
#print(y)
```

```python
plt.scatter(x,y)
plt.show
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
extFor55 = model.predict([[55]])
print(extFor55)
```

```python
print('Slope(m): ',model.coef_)
print('Y-Intercept: ',model.intercept_)

m = model.coef_
c =model.intercept_

print(f'The Linear Regression line equation is : y= {m[0,0]}x+{c[0]}')
```
